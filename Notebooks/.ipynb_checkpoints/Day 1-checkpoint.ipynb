```python
# Sort by patient and admission time
admissions = admissions.sort_values(['subject_id', 'admittime'])

# For each patient, get their NEXT admission date
admissions['next_admittime'] = admissions.groupby('subject_id')['admittime'].shift(-1)

# Convert to datetime
admissions['admittime']      = pd.to_datetime(admissions['admittime'])
admissions['dischtime']      = pd.to_datetime(admissions['dischtime'])
admissions['next_admittime'] = pd.to_datetime(admissions['next_admittime'])

# Days between discharge and next admission
admissions['days_to_readmit'] = (admissions['next_admittime'] - admissions['dischtime']).dt.days

# Target: 1 if readmitted within 30 days, 0 otherwise
admissions['readmitted_30d'] = (admissions['days_to_readmit'] <= 30).astype(int)

# Check the distribution
print(admissions['readmitted_30d'].value_counts())
print(f"\nReadmission rate: {admissions['readmitted_30d'].mean()*100:.1f}%")
```

```python
# Merge patient demographics into admissions
df = admissions.merge(patients, on='subject_id', how='left')

print(f"Shape after merge: {df.shape}")
print(df[['subject_id', 'hadm_id', 'anchor_age', 'gender', 'readmitted_30d']].head())
```

```python
# 1. Length of stay in days
df['los'] = (df['dischtime'] - df['admittime']).dt.days

# 2. Age at admission
df['age'] = df['anchor_age']

# 3. Gender encoded (1 = Male, 0 = Female)
df['gender_male'] = (df['gender'] == 'M').astype(int)

# 4. Discharge location risk tier
high_risk_discharge = ['SKILLED NURSING FACILITY', 'REHAB', 'LONG TERM CARE HOSPITAL']
df['high_risk_discharge'] = df['discharge_location'].isin(high_risk_discharge).astype(int)

# 5. Prior admissions count per patient
df['prior_admissions'] = df.groupby('subject_id').cumcount()

# 6. Medication count per visit
med_count = prescriptions.groupby('hadm_id')['drug'].count().reset_index()
med_count.columns = ['hadm_id', 'num_medications']
df = df.merge(med_count, on='hadm_id', how='left')
df['num_medications'] = df['num_medications'].fillna(0)

# 7. Diagnosis count per visit (comorbidity burden)
diag_count = diagnoses.groupby('hadm_id')['icd_code'].count().reset_index()
diag_count.columns = ['hadm_id', 'num_diagnoses']
df = df.merge(diag_count, on='hadm_id', how='left')
df['num_diagnoses'] = df['num_diagnoses'].fillna(0)

# 8. Abnormal lab count per visit
abnormal_labs = labevents[labevents['flag'] == 'abnormal']
lab_count = abnormal_labs.groupby('hadm_id')['labevent_id'].count().reset_index()
lab_count.columns = ['hadm_id', 'num_abnormal_labs']
df = df.merge(lab_count, on='hadm_id', how='left')
df['num_abnormal_labs'] = df['num_abnormal_labs'].fillna(0)

# Check
features = ['los','age','gender_male','high_risk_discharge',
            'prior_admissions','num_medications','num_diagnoses','num_abnormal_labs']
print(df[features + ['readmitted_30d']].describe())
```

```python
# Sort by admission time — oldest visits first
df = df.sort_values('admittime').reset_index(drop=True)

# Use last 20% of visits as test set
split_idx = int(len(df) * 0.80)

train = df.iloc[:split_idx]
test  = df.iloc[split_idx:]

print(f"Train size : {len(train)} visits | Readmission rate: {train['readmitted_30d'].mean()*100:.1f}%")
print(f"Test size  : {len(test)}  visits | Readmission rate: {test['readmitted_30d'].mean()*100:.1f}%")
```

```python
# Resave train and test with patient IDs
feature_cols = ['subject_id', 'hadm_id', 'los', 'age', 'gender_male', 
                'high_risk_discharge', 'prior_admissions', 'num_medications',
                'num_diagnoses', 'num_abnormal_labs', 'readmitted_30d']

train[feature_cols].to_csv(r'C:\Users\Hp\Desktop\readmission\data\train.csv', index=False)
test[feature_cols].to_csv(r'C:\Users\Hp\Desktop\readmission\data\test.csv', index=False)

print("✅ Saved with patient IDs")
```
